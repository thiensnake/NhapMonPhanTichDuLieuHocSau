```python
import pandas as pd
import numpy as np
```

```python
#1
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ivan", "Jack", "Kelly", "Liam", "Mona", "Nina", "Oscar"],
    "Age": [25, 30, 35, 28, 22, 45, 34, 31, 27, 29, 33, 40, 26, 32, 36],
    "Salary": [50000, 60000, 70000, 55000, 52000, 80000, 72000, 68000, 61000, 59000, 63000, 77000, 53000, 66000, 75000]
}
```

```python
#2 
df = pd.DataFrame(data)
```

```python
df.head(20)
```

```python
#3
dfloc = df[df['Age'] > 28]
print(dfloc)
```

```python
#4 
salary = df['Salary'].mean()
print(salary)
```

```python
#5
groupy = df.groupby('Age')['Salary'].sum()
print(groupy)
```

```python
#6 
sort = df.sort_values(by='Salary', ascending=False)
print(sort)
```

```python
#7
import matplotlib.pyplot as plt

df['Age'].plot(kind='bar', figsize=(10, 6), edgecolor='black')
plt.title('Biểu đồ cột Age')
plt.xlabel('')
plt.ylabel('Age')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#8 
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
df['Salary'].plot(kind='line')
plt.title('Biểu đồ đường Salary')
plt.xlabel('')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
#9
plt.figure(figsize=(5, 5))
df['Age'].value_counts().plot(kind='pie')
plt.title('Biểu đồ tròn Age')
plt.ylabel('')
plt.show()
```

```python
# 10
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'])
plt.title('Biểu đồ phân tán Age và Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
#11 
nan = df.isna().any().any()
print(nan)
```

```python
#12
mean_age = df['Age'].mean()
mean_age_int = int(mean_age)
df.loc[df['Age'] > 30, 'Age'] = mean_age_int
```

```python
df.head(20)
```

```python
#13
df['Age_normalized'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df.head
```

```python
#14
def phanloaituoi(age):
    if age < 30:
        return 'Young'
    elif age <= 60:
        return 'Middle-aged'
    else:
        return 'Old'

df['Age_group'] = df['Age'].apply(phanloaituoi)
df.head
```

```python
# 15
df['Percentage_change'] = df['Salary'].pct_change()
df.head
```

```python
# 16
df_deduped = df.drop_duplicates(subset='Name', keep='first')
df.head
```

```python
#17
df.to_csv('BaiKiemTra.csv', index=False)
```
