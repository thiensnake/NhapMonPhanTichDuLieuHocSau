```python
#5 Thống kê dữ liệu thiếu cho biến T1

data["T1"].isnull().sum()
```

```python
#5 Hiệu chỉnh dữ liệu

data["T1"].fillna(data["T1"].mean(), inplace= False)

    # mean(): được gọi trên cột "T1", trả về giá trị trung bình của cột đó.
    # fillna(): sử dụng để điền các giá trị thiếu bằng một giá trị đã được chỉ định
```

```python
#6 Xử lý lần lượt tất cả các dữ liệu thiếu cho các biến về điểm số
is_float64 = data.dtypes == float  
data_column_list = data.columns[is_float64].tolist()
    # data.dtypes : trả về một Series chứa kiểu dữ liệu của mỗi cột
    # == float: so sánh kiểu dữ liệu của mỗi cột với kiểu dữ liệu float
    # is_float64: là một Series boolean với các giá trị True và False

    # data.columns: Trả về một Index (mảng) chứa các tên cột trong DataFrame data
    # [is_float64]: Dùng để lựa chọn các tên cột tương ứng với các vị trí có giá trị True trong Series 
    # .tolist(): Chuyển đổi Index thành danh sách Python
```

```python
data_column_list
```

```python
# Vòng lặp qua từng cột trong danh sách các cột cần xử lý
for col in data_column_list:
    # Thay thế các giá trị NaN trong cột hiện tại bằng giá trị trung bình của chính cột đó
    # - data[col] truy cập vào cột hiện tại trong DataFrame data
    # - data[col].mean() tính giá trị trung bình của các phần tử trong cột hiện tại
    # - fillna(data[col].mean()) thay thế các giá trị NaN bằng giá trị trung bình tính được
    # - inplace=False nghĩa là trả về một DataFrame mới với các giá trị NaN đã được thay thế,
    #   nhưng không thay đổi trực tiếp DataFrame ban đầu
    data[col].fillna(data[col].mean(), inplace=False)
```

```python
# Hiển thị 5 hàng đầu tiên của DataFrame để kiểm tra các thay đổi
print(data.head())
```

```python
#7: Tạo các biến TBM1, TBM2, TBM3 và tính điểm theo công thức tính
# TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
for i in ["1", "2", "3"]:
    data["TBM"+i]=(data["T"+i]*2+ data["L"+i]+ data["H"+i]+data["S"+i]+data["V"+i]*2+data["X"+i]+data["D"+i]+data["N"+i])/10
```

```python
# Định nghĩa hàm XL để phân loại Điểm Trung Bình Môn (DTBM)
def XL(DTBM):
    # Nếu DTBM nhỏ hơn 5.0, trả về 'Y' (Yếu)
    if DTBM < 5.0:
        return 'Y'
    # Nếu DTBM từ 5.0 đến dưới 6.5, trả về 'TB' (Trung Bình)
    elif 5.0 <= DTBM < 6.5:
        return 'TB'
    # Nếu DTBM từ 6.5 đến dưới 8.0, trả về 'K' (Khá)
    elif 6.5 <= DTBM < 8.0:
        return 'K'
    # Nếu DTBM từ 8.0 đến dưới 9.0, trả về 'G' (Giỏi)
    elif 8.0 <= DTBM < 9.0:
        return 'G'
    # Nếu DTBM từ 9.0 trở lên, trả về 'XS' (Xuất Sắc)
    else:
        return 'XS'
```

```python
# Vòng lặp qua các giá trị "1", "2", "3"
for i in ["1", "2", "3"]:
    # Tạo cột mới "XL"+i trong DataFrame `data` bằng cách áp dụng hàm `XL`
    # vào từng giá trị trong cột "TBM"+i
    # - "TBM"+i là tên cột chứa Điểm Trung Bình Môn (VD: "TBM1", "TBM2", "TBM3")
    # - apply(lambda DTBM: XL(DTBM)) áp dụng hàm `XL` lên từng giá trị DTBM trong cột
    data["XL"+i] = data["TBM"+i].apply(lambda DTBM: XL(DTBM))
```

```python
#9 Tìm giá trị tối thiểu và tối đa của điểm trung bình lớp 10, 11, 12
Diem_min = data[['TBM1', 'TBM2', 'TBM3']].min().min()
    # min(): Phương thức này trả về giá trị tối thiểu cho mỗi cột trong DataFrame con đã trích xuất
    # Có nghĩa là nó sẽ trả về một Series chứa giá trị tối thiểu của "TBM1", "TBM2" và "TBM3"
    
    # min(): Sau khi đã có Series chứa giá trị tối thiểu của các cột, chúng ta gọi lại phương thức .min()
    # để trả về giá trị tối thiểu của tất cả các giá trị tối thiểu đó. Đảm bảo rằng chúng ta có giá trị 
    # tối thiểu nhất trong tất cả các điểm trung bình từ các năm lớp 10, 11 và 12
Diem_max = data[['TBM1', 'TBM2', 'TBM3']].max().max()
```

```python
# Giả sử bạn đã định nghĩa các giá trị Diem_min và Diem_max trước đó
# Ví dụ:
# Diem_min = 0.0  # Điểm tối thiểu có thể có
# Diem_max = 10.0 # Điểm tối đa có thể có

# Vòng lặp qua các giá trị "1", "2", "3"
for i in ["1", "2", "3"]:
    # Tạo cột mới "US_TBM"+i trong DataFrame `data` bằng cách chuẩn hóa
    # giá trị trong cột "TBM"+i từ thang điểm 10 sang thang điểm 4
    # Công thức: ((giá trị - Diem_min) / (Diem_max - Diem_min)) * 4
    data["US_TBM"+i] = ((data["TBM"+i] - Diem_min) / (Diem_max - Diem_min)) * 4



```

```python
# Hiển thị 5 hàng đầu tiên của DataFrame để kiểm tra các thay đổi
data.head()
```

```python
#10 Tạo biến KQXT, mặc định 0 là rớt, nếu kq = 1 là đậu
data['KQXT'] = 0
```

```python
# Tính điểm trung bình của khối A và A1 với điều kiện điểm môn DH1 được nhân đôi,
# sau đó chia cho 4 và kiểm tra nếu kết quả >= 5.0
khoi_A_A1 = ((data['DH1']*2 + data['DH2'] + data['DH3']) / 4) >= 5.0

# Sử dụng phương thức loc để chọn những hàng có KQXT là 'A' hoặc 'A1' và điều kiện khoi_A_A1 là True
# Sau đó, gán giá trị 1 cho cột KQXT của những hàng này
data.loc[data['KQXT'].isin(['A', 'A1']) & khoi_A_A1, 'KQXT'] = 1
```

```python
# Import thư viện numpy để sử dụng np.where
import numpy as np

# Tính điểm trung bình của khối B với điều kiện điểm môn DH2 được nhân đôi,
# sau đó chia cho 4 và kiểm tra nếu kết quả >= 5.0
khoi_B = ((data['DH1'] + data['DH2']*2 + data['DH3']) / 4) >= 5.0

# Sử dụng phương thức loc để chọn những hàng có KQXT là 'B'
# Sau đó, sử dụng np.where để gán giá trị 1 nếu khoi_B là True, ngược lại gán giá trị 0
data.loc[data['KQXT'] == 'B', 'KQXT'] = np.where(khoi_B, 1, 0)
```

```python
# Tính điểm trung bình của các khối khác (C, D, C1, D1),
# sau đó chia cho 3 và kiểm tra nếu kết quả >= 5.0
khoi_khac = ((data['DH1'] + data['DH2'] + data['DH3']) / 3) >= 5.0

# Sử dụng phương thức loc để chọn những hàng có KQXT thuộc các khối 'C', 'D', 'C1', 'D1'
# và điều kiện khoi_khac là True
# Sau đó, gán giá trị 1 cho cột KQXT của những hàng này
data.loc[data['KQXT'].isin(['C', 'D', 'C1', 'D1']) & khoi_khac, 'KQXT'] = 1

```
