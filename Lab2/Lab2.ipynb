```python
import pandas as pd  # Nhập thư viện pandas với bí danh là pd
import numpy as np  # Nhập thư viện numpy với bí danh là np
import matplotlib.pyplot as plt  # Nhập module pyplot từ matplotlib với bí danh là plt
import statsmodels.api as sm  # Nhập thư viện statsmodels với bí danh là sm
```

```python
# Đọc dữ liệu từ tệp CSV và lưu vào DataFrame dftuyensinh,
# sử dụng cột "STT" làm chỉ số (index) cho DataFrame.
dftuyensinh = pd.read_csv("processed_dulieuxettuyendaihoc.csv", index_col="STT")
```

```python
# Sắp xếp DataFrame dftuyensinh theo giá trị của cột 'DH1'
dftuyensinh.sort_values(by='DH1')
```

```python
# Sắp xếp DataFrame dftuyensinh theo giá trị của cột 'GT' (giới tính) và sau đó theo giá trị của cột 'DH1' (điểm thi DH1)
dftuyensinh.sort_values(by=['GT', 'DH1'])
```

```python
# Định nghĩa hàm get_mean để tính giá trị trung bình của một danh sách điểm số.
def get_mean(scores):
    return sum(scores) / len(scores)
```

```python
# Tính tổng các giá trị trong cột 'DH1' của DataFrame dftuyensinh
result = dftuyensinh['DH1'].sum()

# In kết quả ra màn hình
print("\nThe result of applying the function on the dataframe:\n")
print(result)
```

```python
# Định nghĩa hàm Q1 để tính phân vị 25 (Q1) của một mảng hoặc danh sách các số, sử dụng thư viện NumPy.
def Q1(g):
    return np.percentile(g, 25)
```

```python
# Định nghĩa hàm Q2 để tính phân vị 50 (Q2) của một mảng hoặc danh sách các số, sử dụng thư viện NumPy.
def Q2(g):
    return np.percentile(g,50)
```

```python
# Định nghĩa hàm Q3 để tính phân vị 75 (Q3) của một mảng hoặc danh sách các số, sử dụng thư viện NumPy.
def Q3(g):
    return np.percentile(g,75)
```

```python
# Định nghĩa hàm std để tính độ lệch chuẩn của một mảng hoặc danh sách các số, sử dụng hàm np.std() từ thư viện NumPy.
def std(g):
    return np.std(g)
```

```python
# Tạo pivot table từ DataFrame dftuyensinh với chỉ số là cột 'KT' và giá trị là cột 'DH1',
# Tính các giá trị tổng (sum), trung bình (mean), trung vị (median), giá trị nhỏ nhất (min),
# Giá trị lớn nhất (max), độ lệch chuẩn (std), và các phân vị Q1, Q2, Q3 của cột 'DH1'.
pd.pivot_table(data=dftuyensinh, index='KT', values='DH1', aggfunc=['sum', 'mean', 'median', 'min', 'max', std, Q1, Q2, Q3])
```

```python
# Tạo pivot table từ DataFrame dftuyensinh với các chỉ số là cột 'KT' và 'KV' (khu vực),
# và giá trị là cột 'DH1'. Áp dụng các hàm tính toán như tổng (sum), trung bình (mean),
# trung vị (median), giá trị nhỏ nhất (min), giá trị lớn nhất (max), độ lệch chuẩn (std),
# và các phân vị Q1, Q2, Q3 của cột 'DH1' cho mỗi nhóm.
pd.pivot_table(data=dftuyensinh, index= ['KT','KV'], values='DH1', aggfunc=['sum', 'mean', 'median', 'min', 'max', std, Q1, Q2, Q3])
```

```python
# Nhóm dữ liệu trong DataFrame dftuyensinh theo giới tính ('GT') và đếm số lượng mỗi nhóm.
gt = dftuyensinh.groupby('GT')['GT'].agg('count')
```

```python
# Tạo DataFrame mới có tên là cau3 từ dftuyensinh, chỉ chọn những hàng có giới tính là 'M' (nam).
# Nhóm các dòng dữ liệu trong cau3 theo cột 'DT' (dân tộc) và đếm số lượng mỗi nhóm.
cau3 = dftuyensinh[dftuyensinh['GT'] == 'M']
cau3.groupby('DT')['DT'].agg('count')
```
